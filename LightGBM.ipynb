```python
x=df[['mean_radius','mean_texture','mean_area','mean_smoothness']]
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train ,y_valid =train_test_split(x,y,random_state=0)
```

```python
import lightgbm as lgb
model = lgb.LGBMClassifier(colsample_bytree=0.8, reg_alpha=0.1, reg_lambda=0.1,subsample=0.8)
model.fit(x_train,y_train)
```
